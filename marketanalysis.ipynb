```python
#the below helps which column is numerical and which column is categorical
```

```python
cat=[] # List to store categorical columns
num=[] # List to store numerical columns
for column in data.columns:  
    #Checking if the column has more than 10 unique values
    if data[column].nunique()>10:
        #If more than 10 unique values, classify as numerical
        num.append(column)
    else:
        #If 10 or fewer unique values, classify as categorical
        cat.append(column)
```

Display Daily Sales by Day of the Week

```python
data.columns
```

```python
dw_mapping={
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thur',
    4:'Fri',
    5:'Sat',
    6:'Sun'
    
}
```

```python
data['day_of_week'] = data['Date'].dt.dayofweek.map(dw_mapping)
```

```python
data['day_of_week'].value_counts().plot(kind='bar')
```

What Will Be the Highest Months for Sales

```python
month_mapping = {
    1:"Jan",
    2:"Feb",
    3:"March"
    
}
```

```python
data['month']=data['Date'].dt.month.map(month_mapping)
```

```python

data['month'].value_counts().plot(kind='bar')
```
